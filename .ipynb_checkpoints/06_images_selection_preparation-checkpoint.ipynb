### Selecting files
- Number of images for each category has to be the same
- Uninverted images are filtered
- All eyes belong to same side 


```python
outputs = pd.read_csv("d:/project_data/trainLabels.csv")

num_classes = 5
img_paths = [ [] for _ in range (num_classes)]

proceed = 5
level_status = [0,0,0,0,0]

dir_path = 'd:/project_data/train_1'
image_files=os.listdir(dir_path)

while proceed:
    data = selector(image_files, outputs)
    print(data["level"].value_counts())
    print(len(image_files))
    for row in range(len(image_files)):
        level = int(data["level"].iloc[row])
        #print(level, end=' ')
        if level_status[level] < 1200:
            img_paths[level].append(os.path.join(path, image_files[row]))
        elif level_status[level] == 1200:
            proceed-=1
        level_status[level]+=1         
    print(level_status)
```

```python
dir_path = 'd:/project_data/train_1'
files = os.listdir(dir_path)
notchers = []
inverteds = []
for filename in files[:100]:
    path = os.path.join(dir_path, filename)
    print(path)
    if notch_selector(path):
        notchers.append(filename)
    else:
        inverteds.append(filename)
print(f'Notchers\n{notchers}')
print(f'\n\nInverteds\n{inverteds}')
```
