### Selecting files
- Number of images for each category has to be the same
- Uninverted images are filtered
- All eyes belong to same side 


```python
outputs = pd.read_csv("d:/project_data/trainLabels.csv")

dir_path = 'd:\\project_data\\train_1'
image_files=os.listdir(dir_path)
data = selector(image_files, outputs)
#print(data)
#print(len(image_files))

num_classes = 5
img_paths = [ [] for _ in range(num_classes)]

proceed = 5
level_status = [0,0,0,0,0]

class_size = 100
phrase = 'right'

while proceed:    
    for row in range(len(image_files)):
        level = int(data["level"].iloc[row])
        #print(level, end=' ')
        if level_status[level] < class_size:
            path = os.path.join(dir_path, data['image'].iloc[row])+'.jpeg'
            if notch_selector(path) and filename_sorter(path, phrase):#uninverted images only
                img_paths[level].append(path)
                level_status[level]+=1
        if level_status[level] == class_size:
            proceed-=1
            level_status[level]+=1                
    print(level_status)
```
